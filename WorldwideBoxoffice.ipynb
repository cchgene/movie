```python
import re
import time
import json
import pandas as pd
import requests as r
from bs4 import BeautifulSoup
```

```python
tStart = time.time()
WorldwideBoxoffice_rowdata = []
for i in range(1,16725,25):  #16722 titles as of Dec 9 2017
    url = "http://www.worldwideboxoffice.com/index.cgi?top={}&start=1900&finish=2017&order=worldwide&keyword=&links=&popups=".format(i)
    resp = r.get(url)
    WorldwideBoxoffice_rowdata.append(resp.text)

with open("WorldwideBoxoffice_rowdata.json", "w") as f:
    jd = json.dumps(WorldwideBoxoffice_rowdata, ensure_ascii=False, indent=4)
    jd.encode("utf-8")
    f.write(jd)
tEnd = time.time()
print(tEnd-tStart)
```

```python
with open("WorldwideBoxoffice_rowdata.json",'r') as load_f:
    WorldwideBoxoffice_rowdata = json.load(load_f)
```

```python
WorldwideBoxoffice_rowdata
```

```python
movies_box = []
for i in range(int(16725/25)):
    soup = BeautifulSoup(WorldwideBoxoffice_rowdata[i], "html5lib")
    a = str(soup.select("pre")[0])
    b = BeautifulSoup(a,"html5lib")


    for movie in str(b.select("body")[0]).split("\n\n")[1].split("\n"):
        temp = {}
        temp["RANK"] = movie.split("  ")[0].strip()

        prices = re.findall("\$\s+(\d+\.\d)", movie)
        try:
            temp["DOMESTIC"] = float(prices[0].strip())
        except:
            temp["DOMESTIC"] = "0"
        try:
            temp["OVERSEAS"] = float(prices[1].strip())
        except:
            temp["OVERSEAS"] = "0"
        try:
            temp["WORLD"] = float(prices[2].strip())
        except:
            temp["WORLD"] = "0"


    #     <cite><a href="movie.cgi?title=Star%20Wars%3A%20Episode%20I%20%2D%20The%20Phantom%20Menace&amp;year=1999">Star Wars: Episode I - The Phantom Menace</a></cite> (1999)
        temp["TITLES"] = BeautifulSoup(movie, "html5lib").select("cite > a")[0].text.strip()
        temp["YEAR"] = int(re.findall('</cite> \((\d{4})', movie)[0].strip())
        temp["HREF"] = re.findall('href="(.+)">', movie)[0].strip()

        movies_box.append(temp)
    

        
with open("movies_box.json", "w") as f:
    jd = json.dumps(movies_box, ensure_ascii=False, indent=4)
    jd.encode("utf-8")
    f.write(jd)
    
```
